```python
from fastai.vision.all import *
from pathlib import Path
import ipywidgets as widgets
from IPython.display import display

# üîç Buscar autom√°ticamente el archivo export.pkl
model_path = list(Path('.').rglob('export.pkl'))[0]
learn_inf = load_learner(model_path)
print(f"‚úÖ Modelo cargado desde: {model_path}")

# üß± Crear los widgets
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
btn_run = widgets.Button(description="Classify")

# ‚öôÔ∏è Funci√≥n de clasificaci√≥n
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred, pred_idx, probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)

# üéõÔ∏è Mostrar la interfaz
display(btn_upload, btn_run, out_pl, lbl_pred)
```

```python
import os
import shutil
import glob
import urllib.parse

# ===== CONFIGURACI√ìN =====
GITHUB_USERNAME = "juancuestas4"
GITHUB_REPO = "collab"
GITHUB_EMAIL = "cuestasjuan4@hotmail.com"
GITHUB_TOKEN = ""  # Repo p√∫blico, dejar vac√≠o
BRANCH = "main"  # Queremos que la rama principal sea main

# ===== VERIFICAR QUE EL NOTEBOOK EXISTE =====
notebooks = glob.glob("/content/*.ipynb")
if len(notebooks) == 0:
    raise FileNotFoundError("No se encontr√≥ ning√∫n notebook en /content")
NOTEBOOK_NAME = notebooks[0].split("/")[-1]
print("‚úÖ Notebook detectado:", NOTEBOOK_NAME)

# ===== LIMPIAR CLONES PREVIOS =====
%cd /content
if os.path.exists(f"/content/{GITHUB_REPO}"):
    shutil.rmtree(f"/content/{GITHUB_REPO}")

# ===== CONFIGURAR GIT =====
!git config --global user.name "{GITHUB_USERNAME}"
!git config --global user.email "{GITHUB_EMAIL}"

# ===== CLONAR REPO =====
clone_url = f"https://github.com/{GITHUB_USERNAME}/{GITHUB_REPO}.git"
!git clone {clone_url}
%cd {GITHUB_REPO}

# ===== VERIFICAR / CREAR RAMA MAIN =====
branch_list = !git branch
if len(branch_list) == 0:
    # Repo vac√≠o
    !git checkout -b {BRANCH}
    print("Repositorio vac√≠o, creando rama '{}'".format(BRANCH))
else:
    current_branch = branch_list[0].replace("*", "").strip()
    if current_branch != BRANCH:
        !git branch -M {BRANCH}
        print("Rama renombrada a '{}'".format(BRANCH))

# ===== COPIAR NOTEBOOK A LA RA√çZ =====
!cp "/content/{NOTEBOOK_NAME}" .

# ===== COMMIT Y PUSH =====
!git add "{NOTEBOOK_NAME}"
!git commit -m "Subida autom√°tica desde Colab"
!git push -u origin {BRANCH}
print("‚úÖ Notebook '{}' subido correctamente a GitHub en la rama '{}'".format(NOTEBOOK_NAME, BRANCH))

# ===== GENERAR URL DE BINDER =====
notebook_url_encoded = urllib.parse.quote(NOTEBOOK_NAME)
binder_url = "https://mybinder.org/v2/gh/{}/{}/{}?filepath={}".format(
    GITHUB_USERNAME, GITHUB_REPO, BRANCH, notebook_url_encoded
)
print("\nüîó URL de Binder lista para abrir tu notebook:")
print(binder_url)
```
