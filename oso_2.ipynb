```python
from fastai.vision.all import *
from pathlib import Path
import ipywidgets as widgets
from IPython.display import display

# üîç Buscar autom√°ticamente el archivo export.pkl
model_path = list(Path('.').rglob('export.pkl'))[0]
learn_inf = load_learner(model_path)
print(f"‚úÖ Modelo cargado desde: {model_path}")

# üß± Crear los widgets
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
btn_run = widgets.Button(description="Classify")

# ‚öôÔ∏è Funci√≥n de clasificaci√≥n
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred, pred_idx, probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)

# üéõÔ∏è Mostrar la interfaz
display(btn_upload, btn_run, out_pl, lbl_pred)
```

```python
from fastai.vision.all import *
from pathlib import Path
import ipywidgets as widgets
from IPython.display import display
import os

# ===== Paso 1: Subir export.pkl si no est√° =====
if not os.path.exists('/content/export.pkl'):
    print("üìÇ No se encontr√≥ export.pkl. Por favor s√∫belo:")
    from google.colab import files
    uploaded = files.upload()  # Esto abrir√° un selector para subir archivos
    for fn in uploaded.keys():
        if fn.endswith('.pkl'):
            model_path = Path('/content')/fn
            print(f"‚úÖ Archivo {fn} subido correctamente")
else:
    model_path = Path('/content/export.pkl')
    print(f"‚úÖ Archivo export.pkl encontrado en /content")

# ===== Paso 2: Cargar el modelo =====
learn_inf = load_learner(model_path)
print(f"‚úÖ Modelo cargado desde: {model_path}")

# ===== Paso 3: Crear widgets =====
btn_upload = widgets.FileUpload(accept='image/*', multiple=False)
out_pl = widgets.Output()
lbl_pred = widgets.Label()
btn_run = widgets.Button(description="Classify")

# ===== Paso 4: Funci√≥n de clasificaci√≥n =====
def on_click_classify(change):
    if not btn_upload.value:
        lbl_pred.value = "‚ö†Ô∏è Por favor sube una imagen primero."
        return
    # Obtener imagen subida
    uploaded_file = list(btn_upload.value.values())[0]
    img = PILImage.create(uploaded_file['content'])

    # Mostrar la imagen
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))

    # Predecir
    pred, pred_idx, probs = learn_inf.predict(img)
    lbl_pred.value = f'Predicci√≥n: {pred}; Probabilidad: {probs[pred_idx]:.4f}'

btn_run.on_click(on_click_classify)

# ===== Paso 5: Mostrar la interfaz =====
display(btn_upload, btn_run, out_pl, lbl_pred)
```
